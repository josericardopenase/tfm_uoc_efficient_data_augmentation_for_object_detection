# Domain shift: reales vs sintéticas (multi-dataset)

Embeddings CLIP por imagen; KMeans solo en **reales** (elbow); sintéticas proyectadas a esos centroides.

Métricas agregadas por dataset (tabla con `tabulate`). Visualizaciones opcionales por dataset al final.

**Sin red:** define `CLIP_LOCAL_DIR` hacia `./models/clip-vit-base-patch32`.


```python
import os
import random
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import torch
from PIL import Image
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from transformers import CLIPModel, CLIPProcessor

# ====================================================================
# 1. CONFIGURACIÓN DE RUTAS (Corregido y sincronizado con el script plano)
# ====================================================================
DATASETS = {
    "Vessels & Boats": ("datasets_sdb/fernando_ultralytics/images/train", "datasets/sFERNANDO_FLUX29B/images"),
    "KITTI": ("datasets/kitti/images/train", "datasets_sda/sKITTI_FLUXFP8/images"),
    "African Wildlife": ("datasets/african-wildlife/images/train", "datasets/sAFRICANWILDLIFE_FLUXFP8/images"),
    "PASCAL VOC 2012": ("datasets/VOC/images/train", "datasets/sVOC_FLUXFP8/images"),
    # Se añade la ruta absoluta correcta apuntando a la carpeta de imágenes directas generadas por tu script plano
    "Home Objects": ("/home/josericardopenase/Escritorio/datasets/homeobjects-3K/images/train", 
                         "/home/josericardopenase/Escritorio/datasets/sHOMEOBJECTS_FLUXFP8/images"),
    "MSCOCO subset 5k": ("/home/josericardopenase/Escritorio/datasets/coco_subset_5k/images/train", 
                         "/home/josericardopenase/Escritorio/datasets/sCOCO_FLUX/images/train"),
}

MAX_IMAGENES = 300  # Procesar 300 imágenes por grupo

# Cargar CLIP
device = "cuda" if torch.cuda.is_available() else "cpu"
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)

def obtener_embeddings(carpeta_ruta, max_fotos):
    carpeta = Path(carpeta_ruta)
    if not carpeta.is_dir(): 
        print(f"   ❌ Alerta: El directorio no existe o no es válido: {carpeta_ruta}")
        return np.empty((0, 512))
    
    formatos = (".png", ".jpg", ".jpeg", ".webp")
    rutas = [carpeta / f for f in os.listdir(carpeta) if f.lower().endswith(formatos)]
    rutas = rutas[:max_fotos]
    
    if len(rutas) == 0:
        print(f"   ⚠️ Alerta: No se encontraron imágenes válidas en: {carpeta_ruta}")
        return np.empty((0, 512))
        
    embeddings = []
    for r in rutas:
        try:
            img = Image.open(r).convert("RGB")
            inputs = processor(images=img, return_tensors="pt").to(device)
            with torch.no_grad():
                feat = model.get_image_features(**inputs)
                if hasattr(feat, "pooler_output"):
                    feat = feat.pooler_output
                feat = feat / feat.norm(dim=-1, keepdim=True) 
            embeddings.append(feat.cpu().numpy()[0])
        except Exception as e:
            continue  # Saltar imágenes corruptas si las hay
            
    return np.array(embeddings)

# ====================================================================
# 2. PROCESAMIENTO: CLUSTERING Y T-SNE
# ====================================================================
resultados_globales = {}

for nombre, (ruta_real, ruta_sintetica) in DATASETS.items():
    print(f"\nProcesando dataset: {nombre}...")
    
    # 1. Obtener los vectores de las imágenes
    embs_reales = obtener_embeddings(ruta_real, MAX_IMAGENES)
    embs_sinteticos = obtener_embeddings(ruta_sintetica, MAX_IMAGENES)
    
    # Control de flujo explícito para saber qué falló en la consola
    if len(embs_reales) == 0 or len(embs_sinteticos) == 0: 
        print(f"   ⏭️ Saltando {nombre}: Embeddings insuficientes (Reales: {len(embs_reales)}, Sintéticos: {len(embs_sinteticos)})")
        continue

    # 2. KMeans basado en imágenes reales
    kmeans = KMeans(n_clusters=5, random_state=42, n_init="auto")
    clusters_reales = kmeans.fit_predict(embs_reales)
    
    # 3. Predict sobre imágenes sintéticas
    clusters_sinteticos = kmeans.predict(embs_sinteticos)
    
    # 4. Concatenación y t-SNE
    todos_embs = np.vstack([embs_reales, embs_sinteticos])
    coordenadas_2d = TSNE(n_components=2, perplexity=30, random_state=42).fit_transform(todos_embs)
    
    # Guardar resultados mapeados
    resultados_globales[nombre] = {
        "coordenadas": coordenadas_2d,
        "n_reales": len(embs_reales),
        "c_reales": clusters_reales,
        "c_sinteticos": clusters_sinteticos
    }
    print(f"   ✅ Guardado con éxito en resultados_globales.")

print(f"\nProceso finalizado. Claves guardadas en el diccionario: {list(resultados_globales.keys())}")
```

```python
import numpy as np
from scipy.stats import wasserstein_distance

def calcular_distancia_distribuciones(embs_reales, embs_sinteticos):
    """
    Calcula la distancia de Wasserstein entre la distribución de embeddings reales
    y sintéticos. Se calcula por cada dimensión y se promedia.
    """
    # Calculamos la distancia de Wasserstein para cada una de las 512 dimensiones
    # Esto nos da una medida escalar de la divergencia entre los dos sets
    distancias_por_dim = []
    for i in range(embs_reales.shape[1]):
        d = wasserstein_distance(embs_reales[:, i], embs_sinteticos[:, i])
        distancias_por_dim.append(d)
    
    # El valor promedio representa el "gap" global entre dominios
    return np.mean(distancias_por_dim)

wasserstein_por_dataset = {}

for nombre, (ruta_real, ruta_sintetica) in DATASETS.items():
    embs_reales = obtener_embeddings(ruta_real, MAX_IMAGENES)
    embs_sinteticos = obtener_embeddings(ruta_sintetica, MAX_IMAGENES)
    
    if len(embs_reales) > 0 and len(embs_sinteticos) > 0:
        gap = calcular_distancia_distribuciones(embs_reales, embs_sinteticos)
        wasserstein_por_dataset[nombre] = gap
        
# Convertimos el diccionario a un Series para mapearlo fácilmente al DataFrame
s_wasserstein = pd.Series(wasserstein_por_dataset, name="wasserstein_gap")

# Mapeamos los valores de distancia según el dataset
df = df.join(s_wasserstein, on="dataset")

# Verificamos que se haya añadido correctamente
print(df[["dataset", "delta_ap", "wasserstein_gap"]].head())
```

```python
import numpy as np
from sklearn.cluster import KMeans
from scipy.spatial.distance import euclidean
from scipy.stats import entropy

def extraer_metricas_clustering(embs_reales, embs_sinteticos, n_clusters=5):
    """
    Ajusta KMeans en reales, predice en sintéticos y extrae métricas de calidad estructural.
    """
    # 1. Ajustar el modelo solo con la realidad
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init="auto")
    labels_reales = kmeans.fit_predict(embs_reales)
    
    # 2. Ver dónde caen los sintéticos en ese mapa real
    labels_sinteticos = kmeans.predict(embs_sinteticos)
    
    # --- MÉTRICA 1: Centroid Shift (Desplazamiento Semántico) ---
    distancias_centroides = []
    for i in range(n_clusters):
        # Encontrar los puntos que pertenecen al cluster i
        puntos_reales = embs_reales[labels_reales == i]
        puntos_sint = embs_sinteticos[labels_sinteticos == i]
        
        # Solo calcular si hay puntos en ambos (evitar errores de división por cero)
        if len(puntos_reales) > 0 and len(puntos_sint) > 0:
            c_real = puntos_reales.mean(axis=0)
            c_sint = puntos_sint.mean(axis=0)
            distancias_centroides.append(euclidean(c_real, c_sint))
            
    # Promedio del desplazamiento de todos los clusters
    avg_centroid_shift = np.mean(distancias_centroides) if distancias_centroides else np.nan
    
    # --- MÉTRICA 2: Population Skew (Divergencia KL en proporciones) ---
    # Calculamos qué porcentaje de imágenes hay en cada cluster
    prop_reales = np.bincount(labels_reales, minlength=n_clusters) / len(labels_reales)
    prop_sint = np.bincount(labels_sinteticos, minlength=n_clusters) / len(labels_sinteticos)
    
    # Usamos la divergencia de Kullback-Leibler para ver si las proporciones coinciden
    # Sumamos un epsilon (1e-10) para evitar log(0)
    kl_divergence = entropy(prop_reales + 1e-10, prop_sint + 1e-10)
    
    return avg_centroid_shift, kl_divergence

# =========================================================
# Ejecución e integración con tu diccionario
# =========================================================
metricas_clusters = {}

for nombre, (ruta_real, ruta_sintetica) in DATASETS.items():
    embs_reales = obtener_embeddings(ruta_real, MAX_IMAGENES)
    embs_sinteticos = obtener_embeddings(ruta_sintetica, MAX_IMAGENES)
    
    if len(embs_reales) > 0 and len(embs_sinteticos) > 0:
        shift, kl_div = extraer_metricas_clustering(embs_reales, embs_sinteticos, n_clusters=5)
        metricas_clusters[nombre] = {
            "centroid_shift": shift,
            "population_kl": kl_div
        }
        print(f"{nombre} -> Centroid Shift: {shift:.4f} | KL Divergence: {kl_div:.4f}")
```

```python
import os
from sklearn.linear_model import LogisticRegression

# Asegurar que exista la carpeta contenedora para las figuras del paper
os.makedirs("figures", exist_ok=True)

# ====================================================================
# 3. GRÁFICO 1: FILA DE TIPOS (REAL VS SINTÉTICA) con Rectas Estilizadas
# ====================================================================
fig1, axes1 = plt.subplots(1, len(resultados_globales), figsize=(16, 3))

for idx, (nombre, datos) in enumerate(resultados_globales.items()):
    ax = axes1[idx]
    coor = datos["coordenadas"]
    n_real = datos["n_reales"]
    
    # Las primeras 'n_real' filas de las coordenadas son reales, el resto sintéticas
    ax.scatter(coor[:n_real, 0], coor[:n_real, 1], c="#4C72B0", label="Real", alpha=0.6, s=15, zorder=2)
    ax.scatter(coor[n_real:, 0], coor[n_real:, 1], c="#DD8452", label="Synthetic", alpha=0.6, s=15, zorder=2)
    
    # --- REGRESIÓN LOGÍSTICA PARA DATASETS SELECCIONADOS ---
    if nombre in ["KITTI", "PASCAL VOC 2012", "MSCOCO subset 5k", "Vessels & Boats", "African Wildlife", "Home Objects"]:
        # 1. Crear etiquetas (Real=0, Sintética=1)
        y = np.zeros(len(coor))
        y[n_real:] = 1
        
        # 2. Ajustar la regresión lineal sobre las coordenadas 2D
        clf = LogisticRegression(random_state=42)
        clf.fit(coor, y)
        
        # 3. Extraer coeficientes para calcular la recta w0*x + w1*y + b = 0
        w = clf.coef_[0]
        b = clf.intercept_[0]
        
        x_min, x_max = coor[:, 0].min() - 2, coor[:, 0].max() + 2
        x_recta = np.linspace(x_min, x_max, 100)
        y_recta = -(w[0] * x_recta + b) / w[1]
        
        # 4. Dibujar la recta con un tono azul elegante y trazo discontinuo de alta visibilidad
        color_recta = "#1E3A8A"  # Azul oscuro elegante (Midnight Blue)
        ax.plot(x_recta, y_recta, color=color_recta, linestyle="--", linewidth=2.5, label="Boundary", zorder=2, alpha=0.6)
        
        # Acotar los límites dinámicos del plot para evitar distorsiones por la pendiente de la recta
        ax.set_xlim(coor[:, 0].min() - 5, coor[:, 0].max() + 5)
        ax.set_ylim(coor[:, 1].min() - 5, coor[:, 1].max() + 5)
        
        # Añadir el score al título para rigor del paper
        score = clf.score(coor, y) * 100
        ax.set_title(f"{nombre} ({score:.1f}%)", weight="bold")
    else:
        ax.set_title(nombre, weight="bold")
        
    ax.grid(True, linestyle="--", alpha=0.3, zorder=1)
    if idx == 0: ax.legend()

plt.tight_layout()
# GUARDAR GRÁFICO 1
plt.savefig("figures/tsne_real_vs_synthetic.png", dpi=500, bbox_inches='tight', facecolor='white')
plt.show()

# ====================================================================
# 4. GRÁFICO 2: FILA DE CLUSTERS (COLORES POR GRUPO + FORMAS)
# ====================================================================
fig2, axes2 = plt.subplots(1, len(resultados_globales), figsize=(16, 3))

for idx, (nombre, datos) in enumerate(resultados_globales.items()):
    ax = axes2[idx]
    coor = datos["coordenadas"]
    n_real = datos["n_reales"]
    
    # Reales: coloreados por su cluster y con forma de CÍRCULO ('o')
    ax.scatter(coor[:n_real, 0], coor[:n_real, 1], c=datos["c_reales"], cmap="tab10", marker="o", s=20, alpha=0.6)
    
    # Sintéticos: coloreados por su cluster asignado y con forma de CUADRADO ('s')
    ax.scatter(coor[n_real:, 0], coor[n_real:, 1], c=datos["c_sinteticos"], cmap="tab10", marker="s", s=20, alpha=0.6)
    
    ax.set_title(nombre, weight="bold")
    ax.grid(True, linestyle="--", alpha=0.3)

plt.tight_layout()
# GUARDAR GRÁFICO 2
plt.savefig("figures/tsne_semantic_clusters.png", dpi=500, bbox_inches='tight', facecolor='white')
plt.show()
```
