```python
import pandas as pd
DATASET_ROUTE = './datasets/sHOMEOBJECTS_FLUXFP8'
prompts = DATASET_ROUTE + '/prompts.csv'
df = pd.read_csv(prompts)


equivalences = {
    0: "bed",
    1: "sofa",
    2: "chair",
    3: "table",
    4: "lamp",
    5: "television"
}

mapping = {v: k for k, v in equivalences.items()}

df['class'].unique()
```

```python
wrong_count = 0
correct_count = 0

# para estadísticas de cambios
change_stats = {}

for i, x in df.iterrows():
    real_cat = mapping[x['class']]
    index = i

    label = DATASET_ROUTE + "/labels/" + str(index) + ".txt"

    with open(label, "r") as f:
        lines = f.readlines()

    new_lines = []

    for line in lines:
        parts = line.strip().split()

        current_cat = int(parts[0])

        # verificar si estaba bien o mal
        if current_cat == real_cat:
            correct_count += 1
        else:
            wrong_count += 1

            old_name = equivalences[current_cat]
            new_name = equivalences[real_cat]

            key = f"{old_name} -> {new_name}"

            if key not in change_stats:
                change_stats[key] = 0

            change_stats[key] += 1

            print(
                f"Archivo {index}.txt: "
                f"{old_name} ({current_cat}) -> "
                f"{new_name} ({real_cat})"
            )

        # reemplazar categoría
        parts[0] = str(real_cat)

        new_lines.append(" ".join(parts) + "\n")

    # guardar archivo actualizado
    with open(label, "w") as f:
        f.writelines(new_lines)

print("\n===== RESUMEN =====")
print(f"Etiquetas correctas: {correct_count}")
print(f"Etiquetas incorrectas corregidas: {wrong_count}")

print("\n===== CAMBIOS POR CATEGORÍA =====")

for change, count in sorted(change_stats.items(), key=lambda x: x[1], reverse=True):
    print(f"{change}: {count}")
```

```python
import os
import pandas as pd

DATASET_ROUTE = './datasets/sVOC_FLUXFP8'
prompts = DATASET_ROUTE + '/prompts.csv'
df = pd.read_csv(prompts)

classes = {
    0: "aeroplane",
    1: "bicycle",
    2: "bird",
    3: "boat",
    4: "bottle",
    5: "bus",
    6: "car",
    7: "cat",
    8: "chair",
    9: "cow",
    10: "diningtable",
    11: "dog",
    12: "horse",
    13: "motorbike",
    14: "person",
    15: "pottedplant",
    16: "sheep",
    17: "sofa",
    18: "train",
    19: "tvmonitor",
}

# normalizar nombres
def normalize(x):
    return x.lower().replace("_", " ").strip()

# crear mapping normalizado
mapping = {
    normalize(v): k
    for k, v in classes.items()
}

correct = 0
wrong = 0

wrong_examples = []

for _, row in df.iterrows():
    try:
        index = _
    
        # clase esperada desde CSV
        csv_class = normalize(row['class'])
    
        if csv_class not in mapping:
            print(f"Clase desconocida en CSV: {csv_class}")
            continue
    
        expected_id = mapping[csv_class]
    
        label_path = os.path.join(
            DATASET_ROUTE,
            "labels",
            f"{index}.txt"
        )
    
        with open(label_path, "r") as f:
            lines = f.readlines()
    
        for line in lines:
    
            parts = line.strip().split()
    
            current_id = int(parts[0])
    
            if current_id == expected_id:
                correct += 1
            else:
                wrong += 1
    
                wrong_examples.append({
                    "file": f"{index}.txt",
                    "expected": f"{expected_id} ({equivalences[expected_id]})",
                    "found": f"{current_id} ({equivalences[current_id]})"
                })
    except:
        ...

print("\n===== RESULTADO =====")
print(f"Correctas: {correct}")
print(f"Incorrectas: {wrong}")

if wrong_examples:
    print("\n===== EJEMPLOS INCORRECTOS =====")

    for ex in wrong_examples[:20]:
        print(
            f"{ex['file']} | "
            f"esperado: {ex['expected']} | "
            f"encontrado: {ex['found']}"
        )
else:
    print("\n✅ Todas las etiquetas coinciden correctamente")
```
