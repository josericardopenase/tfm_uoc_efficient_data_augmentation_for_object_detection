# Pseudo-etiquetado con SAM3 (equivalente a `groundedsam_labeling.ipynb`)

Por cada imagen: **`get_vision_features` una vez**, luego **`for cls in ALL_CLASSES`** con el texto de esa clase (`grounding_prompt_for`, p. ej. `pedestrian`→`person`). Todas las cajas encontradas van al mismo `.txt` YOLO con el **`class_id` de la clase del forward** (pueden mezclarse varias clases en una imagen). Salidas: `labels_sam3/`, `images_sam3/`.

```python
import os

os.environ.setdefault("CUDA_VISIBLE_DEVICES", "1")

import importlib.util
from pathlib import Path

_cuda_helper = Path.cwd() / "cuda_ld_path.py"
if _cuda_helper.is_file():
    _spec = importlib.util.spec_from_file_location("cuda_ld_path", _cuda_helper)
    _mod = importlib.util.module_from_spec(_spec)
    assert _spec.loader is not None
    _spec.loader.exec_module(_mod)
    _mod.ensure_nvidia_ld_path()

import cv2
import numpy as np
import pandas as pd
import supervision as sv
import torch
from PIL import Image
from tqdm import tqdm
from transformers import Sam3Model, Sam3Processor

DATASET_PATH = "./datasets/sVOC_FLUXFP8"
SAM3_CKPT = "facebook/sam3"
# Umbral de score (sigmoid × presence); 0.35 puede dejar muchas imágenes sin cajas en datos sintéticos
SAM3_INSTANCE_THR = 0.25

df = pd.read_csv(f"{DATASET_PATH}/prompts.csv")
df = df[df["class"] != "misc"]

ALL_CLASSES = sorted(df["class"].unique())
CLASS_TO_ID = {c: i for i, c in enumerate(ALL_CLASSES)}

GROUNDING_PROMPT_BY_CLASS = {
    "pedestrian": "person",
    "Person_sitting": "person",
    "person_sitting": "person",
}


def grounding_prompt_for(label: str) -> str:
    return GROUNDING_PROMPT_BY_CLASS.get(
        label, GROUNDING_PROMPT_BY_CLASS.get(label.lower(), label)
    )


def sam3_boxes_xyxy(res: dict) -> np.ndarray | None:
    """Cajas en xyxy desde post_process (tensor o numpy; si no hay cajas, None)."""
    boxes_t = res.get("boxes")
    if boxes_t is None:
        return None
    if isinstance(boxes_t, np.ndarray):
        arr = np.asarray(boxes_t, dtype=np.float32).reshape(-1, 4)
    elif torch.is_tensor(boxes_t):
        if boxes_t.numel() == 0:
            return None
        arr = boxes_t.detach().float().cpu().numpy().reshape(-1, 4)
    else:
        arr = np.asarray(boxes_t, dtype=np.float32).reshape(-1, 4)
    return arr if arr.size else None


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
processor_sam3 = Sam3Processor.from_pretrained(SAM3_CKPT)
model_sam3 = Sam3Model.from_pretrained(SAM3_CKPT).to(device)
model_sam3.eval()

print(f"Clases ({len(ALL_CLASSES)}): {ALL_CLASSES}")
print(f"Filas / imágenes: {len(df)} | device={device}")
```

`CLASS_TO_ID` sigue el orden de `ALL_CLASSES`. Cada línea del `.txt` es una detección de la **clase del prompt** usado en ese forward, no la clase de la fila del CSV.

```python
path = DATASET_PATH
output_path = os.path.join(path, "labels_sam3")
labeled_images_dir = os.path.join(path, "images_sam3")

os.makedirs(output_path, exist_ok=True)
os.makedirs(labeled_images_dir, exist_ok=True)

box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

for index, row in tqdm(df.iloc[1309:].iterrows(), total=len(df), desc="SAM3 labeling"):
    try:
        image_path = os.path.join(path, "images", f"{index}.png")
        image_bgr = cv2.imread(image_path)
        if image_bgr is None:
            raise FileNotFoundError(f"No se pudo leer: {image_path}")

        h, w = image_bgr.shape[:2]
        image = Image.open(image_path).convert("RGB")
        W_img, H_img = image.size
        assert (W_img, H_img) == (w, h), "Tamaño PIL vs OpenCV inconsistente"

        txt_path = os.path.join(output_path, f"{index}.txt")

        # Una vez visión por imagen; luego un forward SAM3 por cada clase del dataset
        img_inputs = processor_sam3(images=image, return_tensors="pt").to(device)
        with torch.no_grad():
            vision_embeds = model_sam3.get_vision_features(pixel_values=img_inputs.pixel_values)

        orig = img_inputs.get("original_sizes")
        if orig is None:
            target_sizes_list = [[H_img, W_img]]
        else:
            target_sizes_list = orig.tolist() if torch.is_tensor(orig) else list(orig)

        yolo_lines: list[tuple[int, float, float, float, float]] = []
        vis_xyxy: list = []
        vis_class_ids: list[int] = []
        vis_labels: list[str] = []

        for cls_name in ALL_CLASSES:
            text_prompt = grounding_prompt_for(cls_name)
            text_inputs = processor_sam3(text=text_prompt, return_tensors="pt").to(device)
            with torch.no_grad():
                outputs = model_sam3(
                    vision_embeds=vision_embeds,
                    input_ids=text_inputs.input_ids,
                    attention_mask=text_inputs.attention_mask,
                )

            res = processor_sam3.post_process_instance_segmentation(
                outputs,
                threshold=SAM3_INSTANCE_THR,
                mask_threshold=0.5,
                target_sizes=target_sizes_list,
            )[0]

            xyxy = sam3_boxes_xyxy(res)
            if xyxy is None:
                continue
            yolo_cid = CLASS_TO_ID[cls_name]

            for box in xyxy:
                x_min, y_min, x_max, y_max = box
                x_center = ((x_min + x_max) / 2) / w
                y_center = ((y_min + y_max) / 2) / h
                bw = (x_max - x_min) / w
                bh = (y_max - y_min) / h
                yolo_lines.append((yolo_cid, x_center, y_center, bw, bh))
                vis_xyxy.append([x_min, y_min, x_max, y_max])
                vis_class_ids.append(yolo_cid)
                vis_labels.append(cls_name)

        if vis_xyxy:
            det = sv.Detections(
                xyxy=np.asarray(vis_xyxy, dtype=np.float32),
                class_id=np.asarray(vis_class_ids, dtype=np.int32),
            )
            annotated_image = box_annotator.annotate(scene=image_bgr.copy(), detections=det)
            annotated_image = label_annotator.annotate(
                scene=annotated_image,
                detections=det,
                labels=vis_labels,
            )
            with open(txt_path, "w") as f:
                for cid, xc, yc, bw, bh in yolo_lines:
                    f.write(f"{cid} {xc} {yc} {bw} {bh}\n")
        else:
            annotated_image = image_bgr.copy()
            open(txt_path, "w").close()

        cv2.imwrite(os.path.join(labeled_images_dir, f"{index}.png"), annotated_image)
    except Exception as err:
        print(f"Error en índice {index}: {err}")
```
