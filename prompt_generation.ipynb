```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "mistralai/Mistral-7B-Instruct-v0.2"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="cuda:1",
)

inputs = tokenizer(
    "Explain quantum mechanics clearly and concisely.",
    return_tensors="pt"
).to(model.device)

out = model.generate(**inputs, max_new_tokens=150)
print(tokenizer.decode(out[0], skip_special_tokens=True))


```

```python
import pandas as pd
from tqdm import tqdm
import itertools

df = pd.DataFrame({
    "prompt": [],
    "class": [],
    "environment": [],
    "style": [],
    "amount": []
})


environments = [
    "living room",
    "bedroom",
    "kitchen",
    "bathroom",
    "dining room",
    "garage",
    "home office",
    "balcony",
    "hallway",
    "laundry room",
    "basement",
    "garden",
    "patio"
]


styles = [
    "modern",
    "minimalist",
    "rustic",
    "industrial",
    "luxury",
    "vintage",
    "scandinavian",
    "messy",
    "clean",
    "cozy"
]


classes = [
    "laptop",
    "wardrobe",
    "window",
    "door",
    "potted plant",
    "photo frame",

]

amounts = [
    "Single",
    "Two",
    "Three",
    "Multiple",
]

total = len(classes) * len(environments) * len(styles) * len(amounts)
print(total)
exit

for label, env, style, amount in tqdm(
        itertools.product(classes, environments, styles, amounts),
        total=total
    ):

    prompt_text = (
        f"You are a prompt generator. The prompt must be for an amount {amount} "
        f"of a {label} in the {env} with a {style} style, "
        f"dont add tags, just the text prompt. Example: Three chairs in a cozy living room with a modern style. "
        f"Dont add verbs like create, just descriptions. "
        f"JUST OUTPUT THE PROMPT. ALWAYS USE THE WORD {label}. BE AS BRIEF AS POSSIBLE."
    )

    inputs = tokenizer(prompt_text, return_tensors="pt").to(model.device)
    output_ids = model.generate(**inputs, max_new_tokens=50)

    generated_ids = output_ids[0][inputs["input_ids"].shape[-1]:]
    text = tokenizer.decode(generated_ids, skip_special_tokens=True)

    df.loc[len(df)] = [text, label, env, style, amount]
    df.to_csv("home_objects_prompt_2.csv", index=False)
```
