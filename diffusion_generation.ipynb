```python
import time 
import pandas as pd
from tqdm import tqdm

path = "./datasets/sFERNANDO_TUGBOATS_FLUXDEV"

df = pd.read_csv(path + "/prompts.csv")
df = df[df["class"] == "tugboat"]
class_names = df['class'].unique()

for index, row in tqdm(df.iloc[0:].iterrows(), total=len(df), desc="Processing"):
    print(index)    
    other_classes = [name for name in class_names if name != row['class']]
    prompt = row['prompt'].strip().replace("tugboat", "tugboat_xz9")
    pipe_kwargs = {
        "prompt": prompt,
        "height": 640,
        "width": 640,
        "guidance_scale": 3.5,
        "num_inference_steps": 30,
    }
    start = time.time()
    image = pipe(
        **pipe_kwargs,
        generator=torch.manual_seed(0),
    ).images[0]
    device = torch.device('cuda:0')
    #df_attn_maps = get_attn_maps(attn_maps)
    #torch.save(df_attn_maps, f"{path}/images/{index}.pt")
    image.save(f"{path}/images/{index}.png")
    #save_attention_maps(attn_maps, pipe.tokenizer, prompts, base_dir='attn_maps-flux-schnell', unconditional=False)



```
