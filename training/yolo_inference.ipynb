```python
from ultralytics import YOLO
import os

# Cargar pesos entrenados
model = YOLO("./VOC_vainilla_download_yolo26x_b40.pt")

# Carpeta con imágenes
image_dir = "/home/josericardopenase/Escritorio/datasets/sVOC_FLUXFP8/images"

# Carpeta de salida
output_dir = "/home/josericardopenase/Escritorio/datasets/sVOC_FLUXFP8/labels_yolo26x"

# Inferencia + guardar labels YOLO
results = model.predict(
    source=image_dir,
    save=False,
    save_txt=True,      # guarda labels .txt formato YOLO
    save_conf=False,    # opcional: guardar confianza
    project=output_dir,
    name="labels",
    conf=0.25
)
```
