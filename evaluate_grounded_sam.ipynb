```python
from autodistill_grounded_sam import GroundedSAM
from autodistill.detection import CaptionOntology
import os

import torch
import os

# Ensure the correct GPU is targeted before Torch initializes
os.environ["CUDA_VISIBLE_DEVICES"] = "1"


# Map the strings to the names you want the model to use
ontology = CaptionOntology({
    "aeroplane": "aeroplane",
    "bicycle": "bicycle",
    "bird": "bird",
    "boat": "boat",
    "bottle": "bottle",
    "bus": "bus",
    "car": "car",
    "cat": "cat",
    "chair": "chair",
    "cow": "cow",
    "diningtable": "diningtable",
    "dog": "dog",
    "horse": "horse",
    "motorbike": "motorbike",
    "person": "person",
    "pottedplant": "pottedplant",
    "sheep": "sheep",
    "sofa": "sofa",
    "train": "train",
    "tvmonitor": "tvmonitor"
})

model = GroundedSAM(ontology=ontology)
```

```python
import supervision as sv
from supervision.metrics.mean_average_precision import MeanAveragePrecision

# 1. Cargar el dataset
dataset = sv.DetectionDataset.from_yolo(
    images_directory_path="./datasets/VOC/images/val",
    annotations_directory_path="./datasets/VOC/labels/val",
    data_yaml_path="./datasets/VOC/vanilla.yaml"
)

# 2. Inicializar la métrica correctamente
map_metric = MeanAveragePrecision()

for image_name, image, annotations in dataset:
    detections = model.predict(image).with_nms()
    detections = detections[detections.confidence > 0.5]
    map_metric.update(
        predictions=detections,
        targets=annotations
    )

# 4. Calcular el resultado final una vez procesadas todas las imágenes
map_result = map_metric.compute()

# 5. Imprimir resultados
print(f"mAP@50: {map_result.map50}")
print(f"mAP@50-95: {map_result.map50_95}")
```
